Reference Papers:

Latent Diffusion Model: https://arxiv.org/pdf/2112.10752

InST: https://arxiv.org/pdf/2211.13203

This is the challenge part of our project. The referenced work (InST) describes a method can accurately transfer the style
attributes such as semantics, material, object shape, brushstrokes and colors of the references to a natural image with only one style image as guidance.

This is achieved by

1.   applying self-attention mechanism in textual inversion process
2.   add predicted noise (stochastic inversion) to content image and use it as x_start instead of random noise

We aim to optimize texual inversion with the stochastic inversion method on stable diffusion 2 model (imported from https://github.com/Stability-AI/stablediffusion).
